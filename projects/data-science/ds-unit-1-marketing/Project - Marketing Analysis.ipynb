```python
# replace all '?' values in df to null 
df = df.replace({'?': np.nan})
```

```python
# replace incorrect values in 'Sex' column
df = df.replace({
    'Sex':{
        'Male': '1',
        '8': np.nan,
        '9': np.nan
    }
})
```

```python
# drop rows with null values
df = df.dropna()

# again checking to see if df is clean
# df.isnull().any()  
# False for all columns
df.isnull().sum()
```
